```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
df[['2022 Population','2020 Population','2015 Population','2010 Population','2000 Population','1990 Population','1980 Population','1970 Population']] = pd.DataFrame(mm.fit_transform(df[['2022 Population','2020 Population','2015 Population','2010 Population','2000 Population','1990 Population','1980 Population','1970 Population']]))
df.head()
```

```python
from scipy import stats
z_score=np.abs(stats.zscore(df['Growth Rate']))
z_score
```

```python
import seaborn as sns
sns.boxplot(z_score)
```

```python
threshold=z_score.max()
index_val=np.where(z_score >=threshold)
df.iloc[index_val]['Growth Rate']
```

```python
df=df.drop(index=df.iloc[index_val]['Growth Rate'].index)
df
```

```python
mean_population_by_continent = df.groupby("Continent")["2022 Population"].mean().reset_index()

# Create a bar chart for the mean 2022 Population by continent
plt.figure(figsize=(8, 6))
sns.barplot(x="Continent", y="2022 Population", data=mean_population_by_continent, palette="muted")
plt.title("Mean 2022 Population by Continent")
plt.xlabel("Continent")
plt.ylabel("Mean 2022 Population")
plt.show()
```

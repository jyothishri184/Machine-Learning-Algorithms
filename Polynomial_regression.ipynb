```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns #from matplot seaborn is build
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
df
```

```python
X = df.iloc[:,1:2].values
X
```

```python
y = df.iloc[:,2:].values
y
```

```python
from sklearn.linear_model import LinearRegression
model_lr = LinearRegression()
model_lr.fit(X,y)
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,model_lr.predict(X),color='blue')
plt.title('LINEAR REGRESSION')
plt.xlabel('Poistion level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
model_pr2 = PolynomialFeatures(degree=2) #for raising degree
x_poly2 = model_pr2.fit_transform(X)
print(x_poly2) #degree raised x
model_lr2 = LinearRegression()
model_lr2.fit(x_poly2,y) #for model building fit - 2 parameters and for transforming fit_transform - 1 parameter
```

```python
from sklearn.preprocessing import PolynomialFeatures
model_pr3 = PolynomialFeatures(degree=3) #for raising degree
x_poly3 = model_pr3.fit_transform(X)
print(x_poly3) #degree raised x
model_lr3 = LinearRegression()
model_lr3.fit(x_poly3,y) #for model building fit - 2 parameters and for transforming fit_transform - 1 parameter
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,model_lr2.predict(x_poly2),color='blue')
plt.plot(X,model_lr3.predict(x_poly3),color='green')
plt.title('POLYNOMIAL REGRESSION')
plt.xlabel('Poistion level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Real%20estate.csv")
data
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.subplots(figsize=(12,6)) #subplot is for decresing size
sns.heatmap(data.corr(),annot=True,cmap="Greens")
```

```python
sns.displot(data=data,x='Y house price of unit area',kde = True,color='Blue')
```

```python
x = data.drop('Y house price of unit area',axis=1)
x
```

```python
y = data['Y house price of unit area']
y
```

```python
from sklearn.preprocessing import PolynomialFeatures
model_pr = PolynomialFeatures(degree=2,include_bias = False) #for raising degree
#At default bias will be true
```

```python
x.shape
```

```python
poly_feature = model_pr.fit_transform(x)
poly_feature.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(poly_feature,y,test_size = 0.3,random_state=42)
```

```python
print(f'x train {x_train.shape}')
print(f'x test {x_test.shape}')
print(f'y train{y_train.shape}')
print(f'y test {y_test.shape}')
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train,y_train)
y_pred = lr_model.predict(x_test)
```

```python
pd.DataFrame({'Y_TEST': y_test, 'Y_PRED': y_pred, 'Residuals': (y_test-y_pred)})
```

```python
mae = metrics.mean_absolute_error(y_test,y_pred)
mse = metrics.mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print(f'MEAN ABSOLUTE ERROR : {mae}')
print(f'MEAN SQUARED ERROR : {mse}')
print(f'ROOT MEAN SQUARED ERROR : {rmse}')
```
